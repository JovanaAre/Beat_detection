Zadatak 1.   
Napisati funkciju u Pythonu za odredivanje spektralnog fluksa prema
jednačinama datim u uvodu. Funkcija treba da kao ulazne argumente
prima ulazni signal, frekvenciju odmjeravanja, dužinu i tip prozorske
funkcije, veličinu koraka prozora, parametar logaritamske kompresije
γ i dužinu prozora za lokalno usrednjavanje. Izlazni argument treba
da bude spektralni fluks normalizovan tako da maksimalna vrijednost
bude jednaka jedinici.

Implementiraćemo najprije pomoćnu funkciju za računanje lokalne srednje vrijednosti spektralnog fluksa.

```python
#Funkcija za lokalno usrednjavanje (računanje lokalne srednje vrijednosti spektralnog fluksa)
def local_average(x, M):
    L = len(x)
    local_average1 = np.zeros(L)
    for l in range(L):
        m = max(l - M, 0)
        n = min(l + M + 1, L)
        local_average1[l] = (1 / (2 * M + 1)) * np.sum(x[m:n])
    return local_average1
```

Zatim ćemo implementirati traženu funkciju koja računa spektralni fluks, korak po korak, prema zadatim formulama u tekstu domaćeg zadatka.

```python
# Funkcija za određivanje spektralnog fluksa prema zadatim formulama
# Paramatri funkcije su:
# x - ulazni signal, window - tip prozorske funkcije, Fs - frekvencija odmjeravanja
# N - dužina prozorske funkcije, H - veličina koraka prozora
# gamma - parametar logaritamske kompresije i M - dužina prozora za lokalno usrednjavanje
def find_spectral_flux(x, window, Fs=2, N=1024, H=256, gamma=100.0, M=10):
# Računanje STFT ulaznog signala
    X = librosa.stft(x, n_fft=N, hop_length=H, win_length=N, window=window)
    Fs_spectral = Fs / H
    Y = np.log(1 + gamma * np.abs(X))
# Računanje razlike n+1-og i n-tog člana
    Y_diff = np.diff(Y)
# Efekat polutalasnog ispravljanja, gdje se razlika manja od 0 postavlja na 0
    Y_diff[Y_diff < 0] = 0
    spectral_flux = np.sum(Y_diff, axis=0)
    spectral_flux = np.concatenate((spectral_flux, np.array([0.0])))
# Lokalno usrednjavanje
    if M > 0:
        local_average2 = local_average(spectral_flux, M)
        spectral_flux = spectral_flux - local_average2
        spectral_flux[spectral_flux<0] = 0.0
# Normalizacija spektralnog fluksa tako da maksimalna vrijednost bude 1
        max_value = max(spectral_flux)
        if max_value > 0:
            spectral_flux = spectral_flux / max_value
    return spectral_flux, Fs_spectral
```

Zadatak 3.   
Napisati funkciju za detekciju početaka odsviranih nota na osnovu
spektralnog fluksa. Ulazni argumenti treba da budu spektralni fluks i
parametri detektora koje sami odaberete. Izlazni argument treba da
bude lista trenutaka u kojima su detektovani poˇceci odsviranih nota.

Mjesta na kojima dolazi do naglog rasta energije, tj. do pojave lokalnih maksimuma (vršni dijelovi funkcije spektralnog fluksa) su mjesta gdje koja možemo smatrati počecima nota. Zato ćemo u našoj funkciji iskoristiti funkciju iz scipy-a iz signal modula - signal.find_peaks(). Osim što vraća vremenske trenutke odsviranih nota, naša funkcija će izvršiti i iscrtavanje spektralnog fluksa sa označenim izračunatim vršnim dijelovima (kružići na krivoj spektralnog fluksa).

```python
# Funkcija za detekciju početaka odsviranih nota
# Argumenti funkcije noisy i widths se koriste ukoliko je ulazni signal narušen šumom
# Pa u tom slučaju umjesto metode signal.find_peaks(), koristićemo metodu signal.find_peaks_cwt()
# Jer ona vrši i filtriranje ulaznog signala narušenog šumom zbog lakše estimacije početaka nota
# Parametri funkcije su:
# spectral_flux -spektralni fluks, Fs_spectral - frekvencija odmjeravanja spektralnog fluksa
# prominence - potrebna istaknutost vršnih dijelova
# widths - niz širina za izračunavanje cwt (pokriva očekivanu širinu vršnih vrijednosti od interesa)
# noisy - bool vrijednost koja definiše da li će se vršne vrijednosti tražiti pozivom funkcije
# find_peaks() (noisy = False) ili find_peaks_cwt() (noisy = True)
# Razlika je u tome što druga funkcija vrši prvo filtriranje signala narušenog šumom

def notes_detection(spectral_flux, Fs_spectral, prominence, widths, noisy=False):
# peaks -vršni dijelovi(lokalni maksimumi) funkcije spektralnog fluksa
    if noisy is not False:
        peaks = signal.find_peaks_cwt(spectral_flux, widths)
    else:
        peaks, _ = signal.find_peaks(spectral_flux, prominence)
        
    t_vector = np.arange(spectral_flux.shape[0]) / Fs_spectral
# vremenski trenuci detektovanih početaka nota
    notes_time_list = t_vector[peaks]
    plt.plot(notes_time_list, spectral_flux[peaks], 'go')
    return notes_time_list
```

Implementiraćemo i sličnu funkciju za detekciju beat-ova (najvećih vršnih dijelova). Najveće vršne dijelove detektovaćemo proslijeđivanjem odgovarajuće visine (height) funkciji signal.find_peaks(). I to je osnovna ideja detekcije tempa u ovoj realizaciji, jer su otkucaji (beat-ovi) niz opaženih položaja impulsa koji su obično periodično raspoređeni u vremenu, sa približno istom jačinom udara.

```python
# Pomoćna funkcija za detekciju trenutaka u kojima se javljaju otkucaji
# Parametri funkcije su:
# spectral_flux - spektralni fluks, Fs_spectral -frekvencija odmjeravanja spektralnog fluksa
# height - opseg visina funkcije spektralnog fluksa u kojoj se detektuju otkucaji 
# plot - bool vrijednost koja definiše da li će se izvršavati iscrtavanje vertikalnih linija
# u trenucima detektovanih otkucaja (plot = True) ili ne (plot = False)
    
def beats_detection(spectral_flux, Fs_spectral, height, plot = False):
# bets - vektor vršnih dijelova koji predstavljaju beat-ove, najmaksimalnije onsete
    beats, _ = signal.find_peaks(spectral_flux, height)
    t_vector = np.arange(spectral_flux.shape[0]) / Fs_spectral
# vremenski trenuci detektovanih beat-ova
    beats_time_list = t_vector[beats]
    beats_array = np.array(beats_time_list)
    if plot is not False:
        for beat in beats_array: 
            plt.axvline(x=beat, color='red', linestyle='-')
    return beats_time_list
```

Implementiraćemo funkciju za estimaciju tempa polazeći od činjenice da je tempo obrnuto proporcionalan beat periodu (period između 2 uzastopna beat-a (najveća skoka energije signala)). A beat period pronalazimo kao prosiječnu (ili max) vrijednost razlike trenutaka izmedju svaka 2 sukcesivna trenutka u kojima se javlja beat.
Tempo se odnosi na brzinu pulsa i dat je recipročnom vrijednošću otkucaja (beat-ova).

```python
# Funkcija za estimaciju globalnog tempa
# Parametar je:
# beats_array - niz trenutaka u kojim su detektovani otkucaji signala (beat-ovi)
def tempo_estimation(beats_array):
    beats = np.diff(beats_array) #beats je niz u kom su sadržane razlike sukcesivnih elemenata niza beats_array
    #beat_period = sum(beats) / len(beats) # prosječna vrijednost beat perioda
    beat_period = max(beats)
    print("Beat period je:")
    print(beat_period)
    tempo = (1/beat_period) * 60
    tempo_int = np.ceil(tempo)
    return tempo
```

```python
# Funkcija za generisanje sintetičkog spektralnog fluksa
def synth_spectral_flux(bpm, dur=10, Fs=100):
    '''
        Generise sinteticku sekvencu impulsa spektralnog fluksa.
        Ulazni argumenti:
            bpm - tempo u BPM
            dur - trajanje u sekundama
            Fs  - frekvencija odmjeravanja u Hercima
    '''

    L = dur * Fs
    x = np.zeros(L)
    print("(bpm,Fs) =")
    print(bpm, Fs)
    N = 60*Fs // bpm
    #print(N)
    peaks = np.arange(N, L, N)
    x[peaks] = 1
    return x
```

Zadatak 8.     
Implementirati funkciju koja će na osnovu spektralnog fluksa računati
i crtati tempogram. Ulazni argumenti ove funkcije treba da budu ulazni signal, frekvencija odmjeravanja, dužina i tip prozorske funkcije,
veličina koraka prozora i skup vrijednosti mogućih vrijednosti tempa
(predefinisano np.arange(30, 601, 1)). Izlazni argumenti treba da
budu vrijednosti tempograma, vektor za označavanje vremenske ose
(u sekundama) i vektor za označavanje tempo ose (u BPM).


Funkcija za računanje i crtanje tempograma najprije vrši "centriranje" ulaznog signala i dopunjavanje nulama s lijeve i desne strane za vrijednost polovine dužine prozorske funkcije. Nakon toga se vrši računanje matrice tempograma na bazi kratkotrajne Furijeove transformacije. I konačno, vrši se iscrtavanje tempograma, gdje je apscisna osa u funkciji vremena, a ordinata u funkciji tempa (BPM).

```python
# Implementacija funkcije koja računa i crta tempogram
# Parametri funkcije su:
# x - ulazni signal, Fs - frekvencija odmjeravanja, wlen - dužina prozorske funkcije
# window - tip prozorske funkcije, H - veličina koraka prozora
# Tempo - skup mogućih vrijednosti tempa ( np.arange(30, 601, 1))

def tempogram(x, Fs, wlen, window, H, Tempo):
    length = wlen // 2 # cijeli dio količnika
    L = len(x)
    left_len = length
    right_len = length
    total_len = L + left_len + right_len
    # dopunjavanje ulaznog signala nulama
    x_total = np.concatenate((np.zeros(left_len), x, np.zeros(right_len)))
    t_total = np.arange(total_len)
    N = int(np.floor(total_len - wlen) / H) + 1 # np.floor() donji cio dio
    T = len(Tempo)
    X = np.zeros((T, N), dtype=np.complex_) # kompleksna matrica inicijalizovana nulama
# Računanje elemenata tempograma
    for t in range(T):
        # računanje frekvencije na osnovu tempa
        omega = (Tempo[t] / 60) / Fs
        # generisanje kompleksne eksponencijalne funkcije za datu frekvenciju
        complex_exp = np.exp(-2 * np.pi * 1j * omega * t_total)
        #množenje ulaznog signala (dopunjenog nulama) kompleksnom eksponencijalnom funkcijom
        x_exp = x_total * complex_exp
        for n in range(N):
            hop1 = n * H
            hop2 = hop1 + wlen
            X[t, n] = np.sum(window * x_exp[hop1:hop2])
        time = (np.arange(N) * H) / Fs
        BPM = Tempo
    temp = np.abs(X)
    plt.pcolormesh(time, BPM , temp)
    plt.xlabel("Vrijeme [s]")
    plt.ylabel("Tempo [BPM]")
    plt.title("Tempogram")
    return X, time, BPM
```
